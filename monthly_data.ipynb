```python
def find_table(value): 
    #This function takes the value to search for as input and returns a list of tables containing the given value.
    pdf_path = "forecast.pdf"
    tables = read_pdf(pdf_path, pages="all") #Reads the PDF file and stores the tables in a list called tables.

    table_count = 0 #This variable will be used to count the number of tables that contain the given value.
    tables_found = [] # This list will store the tables that contain the given value.
    for table in tables:#Iterates over the list of tables.
        if value in table.to_string(): #Checks if the given value is present in the current table.
            table_count += 1 # If the given value is present, increments the table_count variable by 1.
            tables_found.append(table)# Appends the current table to the tables_found list.
    return tables_found


def print_tables(tables): #This function takes a list of tables as input and prints the tables to the console.
    for i in range(len(tables)): #Iterates over the list of tables.
        print(f"Table {i + 1} :") #Prints the table number.
        print(tables[i]) #Prints the table.
        print("------------")

def get_table_number(tables): #This function takes a list of tables as input and prompts the user to enter a table number.
                              #The function then returns the table number entered by the user.
    table_number = int(input("Enter the table number: ")) # Prompts the user to enter a table number.
    while table_number < 1 or table_number > len(tables):
        #Checks if the table number entered by the user is valid. If the table number is not valid,
        #the function prints an error message and prompts the user to enter the table number again.
        print("Invalid table number.")
        table_number = int(input("Enter the table number: "))

    return table_number #Returns the table number entered by the user.



pdf_path = "forecast.pdf"
value = input("Enter the value to search for in the table: ") # the user to enter a value to search for in the tables.

tables_found = find_table(value) #Calls the find_table() function to find the tables containing the given value.
                                #The results are stored in the list tables_found.

print_tables(tables_found)# Calls the print_tables() function to print the tables in the tables_found list to the console.

table_number = get_table_number(tables_found)#Calls the get_table_number() function to get the table number
#of the table that the user wants to view. The table number is stored in the variable table_number.

print(f"Table {table_number}:") #Prints the table number to the console.
table = tables_found[table_number - 1] #Assigns the table at the specified table number to the variable table.
print(tables_found[table_number - 1]) #Prints the table table to the console.
```
