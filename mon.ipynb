```python
def find_table_by_value(pdf_path, search_value):
    # Open the PDF file
    pdf_file = open(pdf_path, "rb")
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)

    # Iterate through the pages and extract tables
    for page_num in range(pdf_reader.numPages):
        tables = tabula.read_pdf(pdf_path, pages=page_num + 1, multiple_tables=True)

        # Loop through the extracted tables
        for table_num, table in enumerate(tables, start=1):
            # Check if the search_value is in the table
            if (table == search_value).any().any():
                print(f"Table containing '{search_value}' found on Page {page_num + 1}, Table {table_num}:")
                print(table)
                return  # Exit the function once the table is found

    print(f"Table containing '{search_value}' not found in the PDF.")

    # Close the PDF file
    pdf_file.close()

# Replace with the path to your PDF and the value you're searching for
pdf_path = "forecast.pdf"
search_value = "Monthly Data"

find_table_by_value(pdf_path, search_value)
```

```python
import pdfminer
from pdfminer.layout import LTTextBox, LTTextLine, LTTextBoxHorizontal

def extract_tables(pdf_path):

  with open(pdf_path, "rb") as f:
    pdf = pdfminer.PdfFileReader(f)

    pages = pdf.pages

    tables = []
    for page in pages:
      layout = pdfminer.layout.PDFPageLayout(page)

      table = []
      for element in layout:
        if isinstance(element, LTTextBoxHorizontal):
          row = []
          for line in element.lines:
            if isinstance(line, LTTextLine):
              row.append(line.get_text())

          if row:
            table.append(row)

      if table:
        tables.append(table)

    return tables

def main(pdf_path, value):

  tables = extract_tables(pdf_path)

  for table in tables:
    if value in table:
      print(f"Found table on page {table[0][0]}")

if __name__ == "__main__":
  pdf_path = "forecast.pdf"
  value = input("Enter the value to search for in the table: ")

  main(pdf_path, value)
```

```python
#import tabula

class TableOutput:
    def __init__(self):
        self.tables = []

    def get_tables(self):
        return self.tables

    def set_table(self, table):
        self.tables.append(table)


def find_table(values):
    pdf_path = "forecast.pdf"
    tables = read_pdf(pdf_path, pages="all")

    table_count = 0
    tables_found = []
    for table in tables:
        for value in values:
            if value in table.to_string():
                table_count += 1
                tables_found.append(table)

    return tables_found


def print_tables(tables):
    for i in range(len(tables)):
        print(f"Table {i + 1}:")
        print(tables[i])


def get_table_number(tables, output):
    table_numbers = []
    a = int(input("Enter the table number (-1 to quit): "))
    while a != -1:
        table_numbers.append(a)
        a = int(input("Enter the table number (-1 to quit): "))

    for table_number in table_numbers:
        if table_number < 1 or table_number > len(tables):
            print("Invalid table number.")
        else:
            print(f"Table {table_number}:")
            table = tables[table_number - 1]
            print(tables[table_number - 1])
            output.set_table(table)


#if __name__ == "__main__":
output = TableOutput()

pdf_path = "forecast.pdf"
values = ["Mid Continent", "Permian", "Rockies"]

tables_found = find_table(values)

print_tables(tables_found)

get_table_number(tables_found, output)

tables = output.get_tables()

for i in range(len(tables)):
    print(f"Table {i + 1}:")
    print(tables[i])
```
